```python
import pandas as pd # Loading Pandas
import re # Loading regular expressions
```

```python
# Loading the excel sheet 

df = pd.read_excel("C:\\Users\\sahith\\OneDrive\\Documents\\Sample_Longlist_Data.xlsx")
```

```python
# Dropping rows with the null values 

df = df.dropna()
```

```python
df
```

```python
# Dropping the columns which are not needed

df = df.drop(columns = ['Date Added', 'category', 'sellerlink', 'sellerlink-url', 'sellerstorefront-url', 'Count of seller brands', 'Max % of negative seller ratings - last 30 days', 'Max % of negative seller ratings - last 90 days', 'Max % of negative seller ratings - last 12 months', 'Sample brand name', 'Sample Brand URL','seller business name'])
```

```python
df
```

```python
# Sorting the rows based on the product ratings

df = df.sort_values(by ='Hero Product 1 #ratings', ascending = False, ignore_index=True)
```

```python
df
```

```python
# Parsing the seller products whose count = 100,000

df = df[(df.sellerproductcount == '1-16 of over 100,000 results' )]
```

```python
df
```

```python
# Finding the emails of the sellers using Regular Expressions

def find_email(text):
    email = re.findall(r'[\w\.-]+@[\w\.-]+',str(text))
    return ",".join(email)
df['E-mail']=df['sellerdetails'].apply(lambda x: find_email(x))
```

```python
df
```

```python
# Finding the country of the sellers registered 

def Country_code(text):
    Count_code = re.findall((r'US$|DE$|CN$'), str(text))
    return ",".join(Count_code)
df['Country_Code'] = df['businessaddress'].apply(lambda x: Country_code(x))
```

```python
df
```

```python
# Dropping the rows where country is China

df = df.drop(df[df['Country_Code'] == 'CN'].index)
```

```python
df
```

```python
# Dropping the rows whose country is not specified

df = df.drop(df[df['Country_Code'] == ''].index)
```

```python
df
```

```python
# Dropping the rows where emails are  not mentioned

df = df.drop(df[df['E-mail'] == ''].index)
```

```python
df
```

```python
# Finding the contact numbers of the sellers

def find_phone_number(text):
    ph_no = re.findall( r"\d{3}[-]?-\d{3}[-]?-\d{4}|\d{10}|\d{5}[/]\d{4}[-]\d",text) # regular expression for phone numbers listed
    return ",".join(ph_no)
df['number'] = df['sellerdetails'].apply(lambda x: find_phone_number(x))
```
